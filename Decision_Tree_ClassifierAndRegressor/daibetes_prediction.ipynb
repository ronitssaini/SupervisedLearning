```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes
```

```python
dataset=load_diabetes()
```

```python
print(dataset)
```

```python
print(dataset['DESCR'])
```

```python
df_dataset=pd.DataFrame(dataset.data,columns=['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6'])
```

```python
df_dataset
```

```python
result=dataset['target']
```

```python
result
```

```python
X=df_dataset
y=result
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=320)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
params={
    'criterion':['squared_error','friedman_mse','absolute_error'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid=GridSearchCV(estimator=DecisionTreeRegressor(),param_grid=params,cv=5,scoring='neg_mean_squared_error')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
grid.best_index_
```

```python
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
print(mean_absolute_error(y_pred,y_test))
```
