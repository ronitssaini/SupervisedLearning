```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)
#gives scaled values for the dataset, no need to standard scale
```

```python
X
```

```python
pd.DataFrame(X)
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
logistic.fit(X_train,y_train)
```

```python
y_pred=logistic.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
acc=accuracy_score(y_pred,y_test)
print(acc)
matrix=confusion_matrix(y_pred,y_test)
print(matrix)
report=classification_report(y_pred,y_test)
print(report)
```

Hyperparameter tuning and Cross validation

```python
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,1,0.1,0.01]
solver=['newton-cg','lbfgs','liblinear','sag','saga']
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
cv=StratifiedKFold()
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
y_pred1=grid.predict(X_test)
```

```python
grid.best_estimator_
```

```python
grid.best_score_
```

```python
grid.best_index_
```

```python
grid.best_params_
```

```python
acc=accuracy_score(y_pred1,y_test)
print(acc)
matrix=confusion_matrix(y_pred1,y_test)
print(matrix)
report=classification_report(y_pred1,y_test)
print(report)
```

Randomised Search CV

```python
#grid search checks each and every permutation of params but randomised cv checks for params combination randomly, hence fast 
from sklearn.model_selection import RandomizedSearchCV
rand=RandomizedSearchCV(estimator=model,param_distributions=params,cv=5,scoring='accuracy',n_jobs=-1)
```

```python
rand.fit(X_train,y_train)
```

```python
y_pred=rand.predict(X_test)
```

```python
rand.best_estimator_
```

```python
rand.best_params_
```

```python
rand.best_score_
```

```python
acc=accuracy_score(y_pred,y_test)
print(acc)
matrix=confusion_matrix(y_pred,y_test)
print(matrix)
report=classification_report(y_pred,y_test)
print(report)
```

 Logistic Regression on unbalanced dataset

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=10000,n_features=2,n_clusters_per_class=1,n_redundant=0,weights=[0.99],random_state=10,n_classes=2)
```

```python
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,1,0.1,0.01]
solver=['newton-cg','lbfgs','liblinear','sag','saga']
class_weight=[{0:w,1:y} for w in [1,10,50,100] for y in [1,10,50,100]]
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver,class_weight=class_weight)
```

```python
params
```

```python
grid=GridSearchCV(estimator=model,n_jobs=-1,param_grid=params)
```

```python
grid.fit(X_train,y_train)
```

```python
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
acc=accuracy_score(y_pred,y_test)
print(acc)
matrix=confusion_matrix(y_pred,y_test)
print(matrix)
report=classification_report(y_pred,y_test)
print(report)
```
